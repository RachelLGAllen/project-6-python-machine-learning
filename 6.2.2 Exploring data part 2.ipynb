### I'm most interested in how government spending on education as a percentage of GDP in 2019 correlates to these other numbers. Looks like it has a pretty strong negative correlation to papers_cited, and a medium correlation to alumni and highly_cited, and a weak correlation to population size

### Create a scatterplot (or plots) for the variables with the strongest correlations and examine the nature of their relationships

```python
# Create a scatterplot for the "2019" and "papers_cited" columns in seaborn

sns.lmplot(x = '2019', y = 'papers_cited', data = sub)
```

#### Really interesting, there does look to be a negative correlation but also tons of outliers so I wouldn't say there's a clear relationship between those variables

```python
# Create a scatterplot for the "2019" and "alumni" columns in seaborn

sns.lmplot(x = '2019', y = 'alumni', data = sub)
```

#### There is a correlation but no numbers are actually on the line of best fit, they are all outliers, which is certainly interesting 

```python
# Create a scatterplot for the "2019" and "highly_cited" columns in seaborn

sns.lmplot(x = '2019', y = 'highly_cited', data = sub)
```

#### This looks a little better; there's a medium correlation and at least some of the data is on the line of best fit, but again most numbers are outliers 

```python
# Create a scatterplot for the "2019" and "pop" columns in seaborn

sns.lmplot(x = '2019', y = 'pop', data = sub)
```

#### Again lots of outliers but this actually looks like the data that is best actually described by the negative correlation. In other words, the larger the population, the less the country spends on education (as a percentage of GDP) with a few huge outliers right in the middle.
